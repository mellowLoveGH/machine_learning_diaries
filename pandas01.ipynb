```
# https://www.tutorialspoint.com/python_pandas/index.htm

# Series, DataFrame, Panel

# pandas.Series( data, index, dtype, copy)
import pandas as pd
import numpy as np

data = np.array(['a','b','c','d'])

s = pd.Series(data)
print(type(s))
print(s)

s = pd.Series(data,index=[100,101,102,103])
print(type(s))
print(s)

data = {'a' : 0., 'b' : 1., 'c' : 2.}
s = pd.Series(data)
print(s)

data = {'a' : 0., 'b' : 1., 'c' : 2.}
s = pd.Series(data,index=['b','c','d','a', 'e'])
print(s)

s = pd.Series(5, index=[0, 1, 2, 3])
print(s)

s = pd.Series([1,2,3,4,5],index = ['a','b','c','d','e'])
print(s[0])
print(s['d'])
print(s[:3]) # Retrieve the first three elements in the Series
print(s[2:])
print(s[-3:]) # Retrieve the last three elements
print(s[['a','c','d']])
```

```
# pandas.DataFrame( data, index, columns, dtype, copy)
df = pd.DataFrame()
print(df)
print(type(df))

data = [1,2,3,4,5]
df = pd.DataFrame(data)
print(df)
print(type(df))

data = [['Alex',10],['Bob',12],['Clarke',13]]
df = pd.DataFrame(data,columns=['Name','Age'])
print(df)
print(type(df))

data = [['Alex',10],['Bob',12],['Clarke',13]]
df = pd.DataFrame(data,columns=['Name','Age'],dtype=float)
print(df)
print(type(df))

data = {'Name':['Tom', 'Jack', 'Steve', 'Ricky'],'Age':[28,34,29,42]}
df = pd.DataFrame(data)
print(df)
print(type(df))

data = {'Name':['Tom', 'Jack', 'Steve', 'Ricky'],'Age':[28,34,29,42]}
df = pd.DataFrame(data, index=['rank1','rank2','rank3','rank4'])
print(df)
print(type(df))

data = [{'a': 1, 'b': 2},{'a': 5, 'b': 10, 'c': 20}]
df = pd.DataFrame(data)
print(df)
print(type(df))

data = [{'a': 1, 'b': 2},{'a': 5, 'b': 10, 'c': 20}]
df = pd.DataFrame(data, index=['first', 'second'])
print(df)
print(type(df))

data = [{'a': 1, 'b': 2},{'a': 5, 'b': 10, 'c': 20}]
df1 = pd.DataFrame(data, index=['first', 'second'], columns=['a', 'b'])
print(df1)
print(type(df1))
df1 = pd.DataFrame(data, index=['first', 'second'], columns=['a', 'b1'])
print(df1)
print(type(df1))

# Create a DataFrame from Dict of Series
d = {'one' : pd.Series([1, 2, 3], index=['a', 'b', 'c']),
   'two' : pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])}
df = pd.DataFrame(d)
print(df)
print(type(df))

d = {'one' : pd.Series([1, 2, 3], index=['a', 'b', 'c']),
   'two' : pd.Series([1, 2, 3, 4], index=['e', 'f', 'g', 'h'])}
df = pd.DataFrame(d)
print(df)
print(type(df))

print(df ['one'])
print(df ['two'])


d = {'one' : pd.Series([1, 2, 3], index=['a', 'b', 'c']),
   'two' : pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])}
df = pd.DataFrame(d)
df['three']=pd.Series([10,20,30],index=['a','b','c'])
print(df)

df['four']=df['one']+df['three']
print(df)


d = {'one' : pd.Series([1, 2, 3], index=['a', 'b', 'c']), 
   'two' : pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd']), 
   'three' : pd.Series([10,20,30], index=['a','b','c'])}
df = pd.DataFrame(d)
print(df)
del df['one']
print(df)
df.pop('two')
print(df)


d = {'one' : pd.Series([1, 2, 3], index=['a', 'b', 'c']), 
   'two' : pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])}
df = pd.DataFrame(d)
print(df)
print(df.loc['b'])
print(df.iloc[2])
print(df[1:2]) # Multiple rows can be selected using ‘ : ’ operator.


df = pd.DataFrame([[1, 2], [3, 4]], columns = ['a','b'])
print(df)
df2 = pd.DataFrame([[5, 6], [7, 8]], columns = ['a','b'])
df = df.append(df2)
print(df2)
print(df)
df = df.drop(0)
print(df)
```

```
# Python Pandas - Descriptive Statistics
# https://www.tutorialspoint.com/python_pandas/python_pandas_descriptive_statistics.htm
d = {'Name':pd.Series(['Tom','James','Ricky','Vin','Steve','Smith','Jack',
   'Lee','David','Gasper','Betina','Andres']),
   'Age':pd.Series([25,26,25,23,30,29,23,34,40,30,51,46]),
   'Rating':pd.Series([4.23,3.24,3.98,2.56,3.20,4.6,3.8,3.78,2.98,4.80,4.10,3.65])
}
df = pd.DataFrame(d)

print(df)
print(df.sum())
#print(df.sum(1))
print(df.mean())
print(df.std())
print(df.describe())
```

```
# Python Pandas - Function Application
def adder(ele1,ele2):
  return ele1+ele2

df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
print(df)
print(df.pipe(adder,2))
print(df.apply(np.mean))
print(df.apply(np.mean,axis=1)) # row wise
print(df.apply(lambda x: x.max() - x.min()))
print(df['col1'].map(lambda x:x*100))
print(df.applymap(lambda x:x*100))
```

```
# Python Pandas - Reindexing
# reindex_like, rename
N=20

df = pd.DataFrame({
   'A': pd.date_range(start='2016-01-01',periods=N,freq='D'),
   'x': np.linspace(0,stop=N-1,num=N),
   'y': np.random.rand(N),
   'C': np.random.choice(['Low','Medium','High'],N).tolist(),
   'D': np.random.normal(100, 10, size=(N)).tolist()
})
df_reindexed = df.reindex(index=[0,2,5], columns=['A', 'C', 'B'])
print(df)
print()
print(df_reindexed) # 
```

```
# Python Pandas - Iteration
# Do not try to modify any object while iterating
# Iterating is meant for reading and the iterator returns a copy of the original object 
N=20
df = pd.DataFrame({
   'A': pd.date_range(start='2016-01-01',periods=N,freq='D'),
   'x': np.linspace(0,stop=N-1,num=N),
   'y': np.random.rand(N),
   'C': np.random.choice(['Low','Medium','High'],N).tolist(),
   'D': np.random.normal(100, 10, size=(N)).tolist()
   })
for col in df:
  print(col)
print()

# iteritems(), Iterates over each column as key, value pair
#for key,value in df.iteritems():
  #print(key,value)
print()

# iterrows()
for row_index,row in df.iterrows():
  print(row_index,row)
print()

# itertuples()
for row in df.itertuples():
  print(row)
print()

```
